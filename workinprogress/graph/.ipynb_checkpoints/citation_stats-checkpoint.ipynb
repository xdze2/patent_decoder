```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pickle
data = pickle.load( open( "../web/data/patent_infos.pickle", "rb" ) )

print(len(data))
```

```python
# citedby stats, ny years
```

```python
get_date = lambda patent: patent['year'] + patent['month']/12 + patent['day']/365
```

```python
def get_date_citedby(  citedby ):
    dates = []
    for A in citedby:
        dates.append( get_date(data[A] ) )
    
    return dates

def get_delay_citedby(  citedby, date ):
    dates = []
    for A in citedby:
        dates.append( get_date(data[A] )-date )
    
    return dates
```

```python
def get_meanrange( citedby, dateself ):
    dates = get_date_citedby(  citedby )
    
    return (np.array(dates)-dateself).mean() if dates else 0
```

```python
years = []
date_x = []
nber_cited_by = []
meanrange = []
citationdelay = []

for patent in data.values():
    years.append( patent['year'] )
    date_x.append( get_date(patent) )
    nber_cited_by.append( len(patent['citedby']) )
    meanrange.append(get_meanrange( patent['citedby'], get_date(patent)  )  )
    
    citationdelay.extend( get_delay_citedby(  patent['citedby'], get_date(patent) ))  
    
print(len(years))
```

```python
date_x, nber_cited_by = zip( *sorted( zip(date_x, nber_cited_by), key=lambda x:x[0] ) )
```

```python
plt.plot( date_x, range( len(date_x) ), '.', alpha=0.3 )
plt.xlabel('years')
plt.ylabel('nombre de brevets publiés')
```

```python
plt.plot( date_x, nber_cited_by, '.', alpha=0.3 )

plt.xlabel('years')
plt.ylabel('nombre de brevet citant')
```

```python
# brevet qui est cité le plus longtemps ? ...
```

```python
plt.plot( meanrange, nber_cited_by, '.r', alpha=0.3 )
plt.xlabel('moyenne du temps entre le brevet et sa citation (années)')
plt.ylabel('nombre de brevet citant')
```

```python
from scipy.stats import gaussian_kde
```

```python
pdf = gaussian_kde( np.array( sorted(citationdelay) ), 0.2 )
```

```python
x = np.linspace( 0, 125, 300 )
plt.plot( x, pdf.pdf(x) , 'r')

plt.xlabel('temps entre le brevet et sa citation (années)')
plt.ylabel('probabilité')
```

```python
# compare patate et poireau ??
scorelist = []

for patent in data.values():
    
    delay = np.array( get_delay_citedby(  patent['citedby'], get_date(patent) ))  
    score = np.sum( 1-pdf.pdf(delay)  )
    scorelist.append( (patent['patent_number'], score) )
    
print(len(scorelist))
```

```python
sorted( scorelist, key = lambda  x : x[1], reverse=True  )[:5]
```
